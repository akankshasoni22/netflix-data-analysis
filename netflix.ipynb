**3. What is the best time to launch a TV show?**

a. **Find which is the best week to release the Tv-show or the movie.**


b.**which is the best month to release the Tv-show or the movie.**




```python
#a. which is the best week to release the Tv-show or the movie.
data.head()
```

```python
data_movie=data[data.type=='Movie']
data_movie=pd.DataFrame(data_movie)
data_movie
```

```python

# 1. Extract Week Information
data_movie['week_added'] = pd.to_datetime(data_movie['date_added']).dt.isocalendar().week

# 2. Group and Count
movies_per_week = data_movie.groupby('week_added')['title'].count()
print(movies_per_week.sort_values(ascending=False).head(5))

# 3. Find the Maximum
best_week = movies_per_week.idxmax()

print(f"The best week to release a movie is Week {best_week}")


```

```python
x=movies_per_week.index
y=movies_per_week.values
plt.figure(figsize=(12,8))
plt.bar(x,y,color='Green')
```

```python
import pandas as pd

# Assuming 'data' is your DataFrame

# Filter for TV shows
data_TV = data[data['type'] == 'TV Show']

# Convert 'date_added' to datetime, handling errors
data_TV['date_added'] = pd.to_datetime(data_TV['date_added'], format='mixed', errors='coerce')

# Extract week number
data_TV['week_added'] = data_TV['date_added'].dt.isocalendar().week

# Group by week and count titles
tvshow_per_week = data_TV.groupby('week_added')['title'].count()
print(tvshow_per_week.sort_values(ascending=False).head(5))

# Find the week with the most releases
best_week = tvshow_per_week.idxmax()

print(f"The best week to release a TV show is Week {best_week}")
```

**Find which is the best month to release the Tv-show or the movie.
**

```python
#  Find which is the best month to release the Tv-show or the movie.


# Assuming 'data' is your DataFrame

# Convert 'date_added' to datetime objects
data['date_added'] = pd.to_datetime(data['date_added'],format='mixed',errors='coerce')

# Extract month information
data['month_added'] = data['date_added'].dt.month

# Analyze for movies
movie_releases_by_month = data[data['type'] == 'Movie'].groupby('month_added')['title'].count()
print(movie_releases_by_month.sort_values(ascending=False).head(5))
best_month_movies = movie_releases_by_month.idxmax()
print(f"The best month to release a movie is: {best_month_movies}")

# Analyze for TV shows
tvshow_releases_by_month = data[data['type'] == 'TV Show'].groupby('month_added')['title'].count()
print(tvshow_releases_by_month.sort_values(ascending=False).head(5))
best_month_tvshows = tvshow_releases_by_month.idxmax()
print(f"The best month to release a TV show is: {best_month_tvshows}")
```

**4. Analysis of actors/directors of different types of shows/movies.**
a. Identify the top 10 actors who have appeared in most movies or TV shows.


b. Identify the top 10 directors who have appeared in most movies or TV shows.



```python
# prompt: Identify the top 10 actors who have appeared in most movies or TV shows.

# Assuming 'data' is your DataFrame as defined in the provided code.

# a. Top 10 actors
# Explode the 'cast' column to handle multiple actors per title.
cast_exploded = data.assign(cast=data['cast'].str.split(', ')).explode('cast')

# Group by actor and count unique titles, then sort and get top 10.
top_10_actors = cast_exploded.groupby('cast')['title'].nunique().sort_values(ascending=False).head(10)
print("Top 10 actors:\n", top_10_actors)


# b. Top 10 directors
# Explode the 'director' column if needed.
director_exploded = data.assign(director=data['director'].str.split(', ')).explode('director')

top_10_directors = director_exploded.groupby('director')['title'].nunique().sort_values(ascending=False).head(10)
print("\nTop 10 directors:\n", top_10_directors)
```

the movie .**



```python
# prompt: 6. Find After how many days the movie will be added to Netflix after the release of

# Convert 'date_added' to datetime objects
data['date_added'] = pd.to_datetime(data['date_added'], format='mixed', errors='coerce')

# Calculate the difference between 'date_added' and 'release_year'
data['days_to_netflix'] = (data['date_added'] - pd.to_datetime(data['release_year'], format='%Y')).dt.days

# Find the mode of the difference
mode_days = data['days_to_netflix'].mode()[0]

print(f"The mode of the days to add a movie to Netflix after release is: {mode_days} days")
### 334 days
```
